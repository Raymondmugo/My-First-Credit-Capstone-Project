```python
# 1. Search for columns that might be dates
date_candidates = [col for col in working_data.columns if '_d' in col.lower() or 'date' in col.lower()]
print("Potential date columns found:")
print(date_candidates)

# 2. Check specifically for 'issue_d' and convert it
if 'issue_d' in working_data.columns:
    print("\nConfirmed: 'issue_d' exists.")
    
    # Convert from string to a proper pandas datetime object
    working_data['issue_d'] = pd.to_datetime(working_data['issue_d'])
    
    # Check the timeline of your dataset
    print(f"Oldest Loan: {working_data['issue_d'].min()}")
    print(f"Newest Loan: {working_data['issue_d'].max()}")
else:
    print("\n'issue_d' not found. We might need to check the column list manually.")
```

```python
# Keep only loans with a definitive outcome
valid_statuses = ['Fully Paid', 'Charged Off', 'Default']
working_data = working_data[working_data['loan_status'].isin(valid_statuses)].copy()

# Create the binary target (1 for default, 0 for fully paid)
working_data['bad_loan'] = np.where(working_data['loan_status'].isin(['Charged Off', 'Default']), 1, 0)
```

```python
# Total exposure of defaulted loans
total_written_off = working_data[working_data['bad_loan'] == 1]['loan_amnt'].sum()

print(f"Total Amount Written Off: ${total_written_off:,.2f}")
```

```python
# Default rate by Home Ownership
home_risk = working_data.groupby('home_ownership')['bad_loan'].mean().sort_values(ascending=False) * 100
print("Default Rate by home ownership  (%):\n", home_risk)
```

```python
# 1. Identify numeric and categorical columns
numeric_cols = working_data.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = working_data.select_dtypes(include=['object']).columns

# 2. Impute numeric columns with the median
for col in numeric_cols:
    if working_data[col].isnull().sum() > 0:
        working_data[col] = working_data[col].fillna(working_data[col].median())

# 3. Impute categorical columns with 'Unknown'
for col in categorical_cols:
    if working_data[col].isnull().sum() > 0:
        working_data[col] = working_data[col].fillna('Unknown')

# 4. The Final Sanity Check
total_missing = working_data.isnull().sum().sum()
print(f"Total missing values remaining in the entire dataset: {total_missing}")
```

```python
# 1. Isolate the object columns
categorical_cols = working_data.select_dtypes(include=['object']).columns

# 2. Check the number of unique values in each text column
cardinality = working_data[categorical_cols].nunique().sort_values(ascending=False)
print("Unique values per categorical column:\n", cardinality)
```

```python
# 1. The Hitlist: Drop cols with high Cardinality
cols_to_drop = [
    'emp_title', 'title', 'zip_code', 'addr_state', 'earliest_cr_line', 
    'last_credit_pull_d', 'last_pymnt_d', 'sub_grade', 'loan_status', 'pymnt_plan','debt_settlement_flag_N','debt_settlement_flag_Y'
]
# errors='ignore' ensures it won't crash if a column was already dropped earlier
working_data = working_data.drop(columns=cols_to_drop, errors='ignore')

# 2. Isolate the numerical text columns
categorical_survivors = working_data.select_dtypes(include=['object']).columns
print(f"Features to be One-Hot Encoded: {list(categorical_survivors)}")

# 3. One-Hot Encode (Translate text to binary 0s and 1s)
# drop_first=True is a vital statistical trick to avoid "Multicollinearity" 
working_data = pd.get_dummies(working_data, columns=categorical_survivors, drop_first=True)

# 4. The Final Check
print(f"\nNew ML-Ready Dataset Shape: {working_data.shape}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 1. Isolate Features (X) and Target Variable (y)
# Added 'income_band' to the drop list to fix the interval casting error
X = working_data.drop(columns=['bad_loan', 'issue_d', 'issue_month', 'income_band'], errors='ignore')
y = working_data['bad_loan']

# 2. Perform the Stratified Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

print("--- Data Split Summary ---")
print(f"Total Features Being Modeled: {X_train.shape[1]}")
print(f"Training Observations (X_train): {X_train.shape[0]:,}")
print(f"Testing Observations (X_test):   {X_test.shape[0]:,}")
print(f"Training Default Rate: {y_train.mean() * 100:.2f}%")
print(f"Testing Default Rate:  {y_test.mean() * 100:.2f}%\n")

# 3. Prevent Magnitude Bias with StandardScaler
scaler = StandardScaler()

# Fit only on the training data, then transform both
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("--- Scaling Sanity Check ---")
print(f"Scaled Training Mean (Should be ~0): {X_train_scaled.mean().round(2)}")
print(f"Scaled Training Std Dev (Should be 1): {X_train_scaled.std().round(2)}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 1. Initialize the model
log_reg = LogisticRegression(max_iter=1000, n_jobs=-1, random_state=42)

# 2. Train the model (The Algorithm is studying the textbook)
print("Training Logistic Regression model... (This may take a minute or two!)")
log_reg.fit(X_train_scaled, y_train)

# 3. Make predictions on the Test Set (The Final Exam)
print("Making predictions on the test set...")
y_pred = log_reg.predict(X_test_scaled)

# 4. Calculate the baseline accuracy
baseline_accuracy = accuracy_score(y_test, y_pred)
print(f"\nBaseline Accuracy: {baseline_accuracy * 100:.2f}%")
```

```python
# 1. The Financial Leakage Hitlist
financial_leakers = [
    'out_prncp', 'out_prncp_inv', 'total_pymnt', 'total_pymnt_inv', 
    'total_rec_prncp', 'total_rec_int', 'total_rec_late_fee', 
    'recoveries', 'collection_recovery_fee', 'last_pymnt_amnt'
]

# Drop them from our feature set
X_clean = X.drop(columns=financial_leakers, errors='ignore')

# 2. Re-Split the Data
X_train, X_test, y_train, y_test = train_test_split(
    X_clean, y, test_size=0.20, random_state=42, stratify=y
)

# 3. Re-Scale the Data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"Features remaining after deep purge: {X_train_scaled.shape[1]}")

# 4. Re-Train the Baseline Model
log_reg = LogisticRegression(max_iter=1000, n_jobs=-1, random_state=42)
print("Retraining Logistic Regression without the time machines...")
log_reg.fit(X_train_scaled, y_train)

# 5. The Moment of Truth
y_pred = log_reg.predict(X_test_scaled)
true_accuracy = accuracy_score(y_test, y_pred)
print(f"\nTrue Baseline Accuracy: {true_accuracy * 100:.2f}%")
```

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 1. Generate the Classification Report
# This shows Precision, Recall, and F1-Score for both Good (0) and Bad (1) loans
print("--- Classification Report ---")
print(classification_report(y_test, y_pred))

# 2. Build the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# 3. Visualize it for the Capstone Portfolio
fig, ax = plt.subplots(figsize=(8, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Good Loan (0)', 'Default (1)'])
disp.plot(cmap='Blues', ax=ax, values_format='d')

plt.title('Credit Risk Confusion Matrix')
plt.show()

# 4. Calculate the specific Financial Errors
true_negatives = cm[0][0]   # Correctly predicted Good Loans
false_positives = cm[0][1]  # Predicted Bad, but were actually Good (Lost Interest)
false_negatives = cm[1][0]  # Predicted Good, but actually Defaulted (Lost Principal!)
true_positives = cm[1][1]   # Correctly predicted Defaults (Money Saved!)

print("\n--- Business Error Breakdown ---")
print(f"Danger - False Negatives (Missed Defaults): {false_negatives:,} loans")
print(f"Caution - False Positives (Lost Customers): {false_positives:,} loans")
```

```python
# 1. Initialize the model with Cost-Sensitive Weights
# class_weight='balanced' forces the model to treat the 20% minority class (Defaults) 
# as equally important to the 80% majority class.
balanced_log_reg = LogisticRegression(max_iter=1000, n_jobs=-1, random_state=42, class_weight='balanced')

# 2. Retrain the model
print("Retraining the mathematically balanced model...")
balanced_log_reg.fit(X_train_scaled, y_train)

# 3. Generate the new predictions
y_pred_balanced = balanced_log_reg.predict(X_test_scaled)

# 4. Build the new Confusion Matrix
cm_balanced = confusion_matrix(y_test, y_pred_balanced)

# 5. Calculate the new Financial Errors
new_false_negatives = cm_balanced[1][0]  
new_false_positives = cm_balanced[0][1]  

print("\n--- NEW Business Error Breakdown ---")
print(f"Danger - False Negatives (Missed Defaults): {new_false_negatives:,} loans")
print(f"Caution - False Positives (Lost Customers): {new_false_positives:,} loans")
print(f"\nWe reduced Missed Defaults by: {43331 - new_false_negatives:,} loans!")
```

```python
from xgboost import XGBClassifier

# 1. Calculate the exact imbalance ratio for XGBoost
# We count how many 0s (Good) and divide by how many 1s (Bad) in the training set
ratio = float(np.sum(y_train == 0)) / np.sum(y_train == 1)
print(f"Calculated scale_pos_weight: {ratio:.2f}")

# 2. Initialize XGBoost
# max_depth controls how complex the trees get (3-6 is usually safe to prevent overfitting)
# learning_rate controls how aggressively it corrects its mistakes
xgb_model = XGBClassifier(
    n_estimators=200,
    max_depth=4,
    learning_rate=0.1,
    scale_pos_weight=ratio, 
    random_state=42,
    n_jobs=-1
)

# 3. Train the Model
xgb_model.fit(X_train_scaled, y_train)

# 4. Make Predictions
y_pred_xgb = xgb_model.predict(X_test_scaled)

# 5. The Final Financial Matrix
cm_xgb = confusion_matrix(y_test, y_pred_xgb)

xgb_false_negatives = cm_xgb[1][0]  
xgb_false_positives = cm_xgb[0][1]  

print("\n--- XGBOOST Business Error Breakdown ---")
print(f"Danger - False Negatives (Missed Defaults): {xgb_false_negatives:,} loans")
print(f"Caution - False Positives (Lost Customers): {xgb_false_positives:,} loans")
```

```python
import shap

# 1. Initialize the SHAP explainer specifically for XGBoost
explainer = shap.TreeExplainer(xgb_model)

# 2. Calculate SHAP values for the test set
X_sample = X_test_scaled[:5000]
shap_values = explainer.shap_values(X_sample)

# 3. Generate the Summary Plot
shap.summary_plot(shap_values, X_sample, feature_names=X_clean.columns)
```

```python
# 1. We need the specific explanation object for the waterfall plot
explainer_obj = explainer(X_sample)

# 2. Plot the exact breakdown for the very first applicant in our test set (Index 0)
shap.plots.waterfall(explainer_obj[0], max_display=10)
```
